```python
import os
import os.path
import pandas as pd
import matplotlib.pyplot as plt

set100 = os.listdir("""[SETSmart] SET100 Prices 2015-2019/SET100_Data""")
```

```python
dataFolder = "YahooFinance"

df_all = pd.DataFrame

data = {}

for series in os.listdir(dataFolder):
    symbol = series.replace('.csv', '')
    if symbol in set100:
        df = pd.read_csv(os.path.join(dataFolder,series), index_col=0)
        # print(symbol)
        data[symbol] = df
len(data)
```

```python
df = pd.DataFrame()
for s, d in data.items():
    df[s] = d['Open']
df
```

```python
df.plot()
```

```python
# Plot each symbol
for series in os.listdir(dataFolder):
    symbol = series.replace('.csv', '')
    if symbol in set100:
        df = pd.read_csv(os.path.join(dataFolder,series), index_col=0)
        print(symbol)
        df['Open'].plot()
```

```python
def each_stat(symbol):
    df = data[symbol]
    df[['Open','Close']].plot()
each_stat("PTT")
```

```python
ptt = data['PTT']
ptt['Previous'] = ptt['Open'].shift()
ptt
```

```python
each_stat('SCC')
```

```python
ptt['%Change'] = (ptt['Open'] - ptt['Previous'])/ptt['Previous']
ptt
```

```python
def stock_return(symbol):
    """Compute return. Drop first record"""
    df = data[symbol]
    df['Previous'] = df['Open'].shift() # shift one day
    df['%Change'] = (df['Open'] - df['Previous'])/df['Previous'] * 100
    return df[1:]
```

```python
df = stock_return("SCC")
df
```

```python


df_all['SCB'].plot(label='Daily', alpha=0.5, figsize=(20,5))
df_all['SCB'].resample('1M').pad().plot(label='Monthly - pad')
df_all['SCB'].resample('1M').bfill().plot(label='Monthly - bfill')
df_all['SCB'].resample('1M').asfreq().plot(label='Monthly - asfreq')
plt.legend()
```

```python
# Resampling documentation
#  https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#resampling

df_all['SCB']['2015-01-01':'2019-01-01']

df_all['SCB']['2015-01-01':'2019-01-01'].plot(label='Daily', alpha=0.5, figsize=(20,5))
df_all['SCB']['2015-01-01':'2019-01-01'].resample('1M').pad().plot(label='Monthly - pad')
df_all['SCB']['2015-01-01':'2019-01-01'].resample('1M').bfill().plot(label='Monthly - bfill')
df_all['SCB']['2015-01-01':'2019-01-01'].resample('1M').asfreq().plot(label='Monthly - asfreq')
df_all['SCB']['2015-01-01':'2019-01-01'].resample('1M').mean().plot(label='Monthly - mean')
df_all['SCB']['2015-01-01':'2019-01-01'].resample('1M').ffill().plot(label='Monthly - ffill')
plt.legend()
```
