# All stock data from Yahoo

```python
import datetime as dt

start = dt.datetime(2014,1,1)

pdr.get_data_yahoo('AOT.BK',start=start)
```

```python
stocks = pd.read_csv("SET_symbols.csv")
stocks
```

```python
for s in stocks['symbol'].values:
    print(s)
    try:
        data = pdr.get_data_yahoo('%s.BK' % s,start=start)
        outfile = "YahooFinance/%s.csv" % s
        data.to_csv(outfile)
    except Exception as e:
        print("Error: %s %s" % (s, str(e)))
```
