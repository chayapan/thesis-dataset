```python
data = {}
```

```python
# Date series

r = 6
j = 3   # Min data
k = 184 # Max data
s = []
for i in range(j,k):
    print(worksheet.cell(r,i).value)
    s.append(worksheet.cell(r,i).value.strip())
data['Period'] = s
```

# Series

Bangkok Interbank Offered Rate (BIBOR) 2/   1-month    25

Thai Baht Implied Interest Rate 2/     1-month   32

T-Bill & Government Bond Yield 6/     1-month   42

```python
# Bangkok Interbank Offered Rate (BIBOR) 1-month 

r = 25
j = 3   # Min data
k = 184 # Max data
s = []
for i in range(j,k):
    print(worksheet.cell(r,i).value)
    s.append(float(worksheet.cell(r,i).value))
data['Bangkok Interbank Offered Rate (BIBOR) 1-month'] = s
```

```python
# Thai Baht Implied Interest Rate 1-month 

r = 32
j = 3   # Min data
k = 184 # Max data
s = []
for i in range(j,k):
    print(worksheet.cell(r,i).value)
    s.append(float(worksheet.cell(r,i).value))
data['Thai Baht Implied Interest Rate 1-month'] = s
```

```python
# T-Bill & Government Bond Yield 6/     1-month  

r = 42
j = 3   # Min data
k = 184 # Max data
s = []
for i in range(j,k):
    print(worksheet.cell(r,i).value)
    s.append(float(worksheet.cell(r,i).value))
data['T-Bill & Government Bond Yield 1-month'] = s
```
