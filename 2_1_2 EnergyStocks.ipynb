```python
import requests
from lxml import html

# Get Sector Symbols
url = "https://www.settrade.com/C04_08_stock_sectorcomparison_p1.jsp?txtSymbol=EGCO"
r = requests.get(url)
print(r.status_code, r.encoding)
tree = html.fromstring(r.content)
tree

# dir(tree.xpath('//a')[4])
```

```python
symbols = {}

# Parse symbols from table in the web page.

for i,e in enumerate(tree.xpath('//table//td')):
    try:
        print("{} {}".format(str(i), e.text_content()))
        if 'text-left' in e.attrib.values():
            print(e.text_content())
            symbols[e.text_content()] = e.text_content()
    except:
        print("{} Error".format(i))

symbols
```

```python
# Historical Trading download from Setsmart
# Use cookie from web browser.   document.cookie

"""
http://setsmart.th.rhbtradesmart.com/historicalTrading.html?lstDisplay=T&symbol=IFEC&lstMethod=TR&decorator=excel&showBeginDate=13%2F11%2F2013&lstPeriod=D&endDate=12%2F11%2F2019&beginDate=13%2F11%2F2013&quickPeriod=6Y&showEndDate=12%2F11%2F2019

http://setsmart.th.rhbtradesmart.com/historicalTrading.html?lstDisplay=T&symbol=IFEC&lstMethod=TR&decorator=excel&showBeginDate=13%2F11%2F2013&lstPeriod=D&endDate=12%2F11%2F2019&beginDate=13%2F11%2F2013&quickPeriod=6Y&showEndDate=12%2F11%2F2019
"""


def fetch_data(symbol):
    url = "http://setsmart.th.rhbtradesmart.com/historicalTrading.html?lstDisplay=T&symbol={}&lstMethod=TR&decorator=excel&showBeginDate=13%2F11%2F2013&lstPeriod=D&endDate=12%2F11%2F2019&beginDate=13%2F11%2F2013&quickPeriod=6Y&showEndDate=12%2F11%2F2019"
    # cookie = """__utmz=198114410.1567759007.2.2.utmcsr=google|utmccn=(organic)|utmcmd=organic|utmctr=(not%20provided); __utma=198114410.1989881950.1564030386.1572491806.1573634807.9; __utmc=198114410; __utmt=1; __utmb=198114410.1.10.1573634807; JSESSIONID=7C6DBD866F8C290BE596B9A3013E2039; TS01bd6b93=01033e365a886a2e4dd5c4d287303f0f913b66e732b54b64b95270a78837d59a182481211e640c81c1ed68919dac7e7a1865953668"""
    cookie = """__utmz=198114410.1567759007.2.2.utmcsr=google|utmccn=(organic)|utmcmd=organic|utmctr=(not%20provided); __utmc=198114410; JSESSIONID=7C6DBD866F8C290BE596B9A3013E2039; __utma=198114410.1989881950.1564030386.1573634807.1573637078.10; __utmb=198114410.1.10.1573637078; TS01bd6b93=01033e365ab44f1ce056036b13d351258bbe2890eba3ccd31d70e7b4fc41b37b7d04dd9a09294eef459c615feb746efdd575e77bdf"""
    url = url.format(symbol)
    
    jar = requests.cookies.RequestsCookieJar()
    cookie_items = cookie.split(';')
    for i in cookie_items:
        k = i.split('=')[0]
        v = "".join(i.split('=')[1:])
        jar.set(k, v, domain='setsmart.th.rhbtradesmart.com', path='/')

    # jar.get_dict()

    r = requests.get(url, cookies=jar)
    print("""Historical Trading""", (r.status_code, r.encoding))
    print(r.url)
    print(r.headers)
    with open("%s/HistoricalTrading.xls" % symbol, "wb") as f:
        f.write(r.content)
    print("Content saved.")
    # Use pandas instead of xlrd. XLRD receive error.
    df_list = pd.read_html('%s/HistoricalTrading.xls' % symbol)
    df = pd.DataFrame(df_list[0])
    print(df)
    
    
    # Balance Sheet
    fs_bs = """http://setsmart.th.rhbtradesmart.com/financialstatement.html?lstEndYear=2019&lstDisplay=T&lstStatementType=B&lstBeginPeriod=Q1&symbol={}&decorator=excel&lstAccountType=U&submit.y=13&submit.x=12&lstCompareType=Q&lstEndPeriod=Q2&lstBeginYear=2014"""
    url = fs_bs.format(symbol)
    r = requests.get(url, cookies=jar)
    print("""Balance Sheet""", r.status_code, r.encoding)
    print(r.headers)
    with open("%s/BalanceSheet.xls" % symbol, "wb") as f:
        f.write(r.content)
    df_list = pd.read_html('%s/BalanceSheet.xls' % symbol)
    df = pd.DataFrame(df_list[0])
    print(df)

    # Income Statement
    fs_is = """http://setsmart.th.rhbtradesmart.com/financialstatement.html?lstEndYear=2019&lstDisplay=T&lstStatementType=I&lstBeginPeriod=Q1&symbol={}&decorator=excel&lstAccountType=U&submit.y=12&submit.x=26&lstCompareType=Q&lstEndPeriod=Q2&lstBeginYear=2014"""
    url = fs_is.format(symbol)
    r = requests.get(url, cookies=jar)
    print("""Income Statement""", r.status_code, r.encoding)
    print(r.headers)
    with open("%s/IncomeStatement.xls" % symbol, "wb") as f:
        f.write(r.content)
    df_list = pd.read_html('%s/IncomeStatement.xls' % symbol)
    df = pd.DataFrame(df_list[0])
    print(df)
    
    # Cashflow Statement
    fs_cf = """http://setsmart.th.rhbtradesmart.com/financialstatement.html?lstEndYear=2019&lstDisplay=T&lstStatementType=C&lstBeginPeriod=Q1&symbol={}&decorator=excel&lstAccountType=U&submit.y=8&submit.x=18&lstCompareType=Q&lstEndPeriod=Q2&lstBeginYear=2014"""
    url = fs_is.format(symbol)
    r = requests.get(url, cookies=jar)
    print(r.status_code, r.encoding)
    print(r.headers)
    with open("%s/CashflowStatement.xls" % symbol, "wb") as f:
        f.write(r.content)
    df_list = pd.read_html('%s/CashflowStatement.xls' % symbol)
    df = pd.DataFrame(df_list[0])
    print(df)
    
    return df_list


for s in symbols:
    try:
        fetch_data(s)
    except Exception as e:
        print("Error for {}: {}".format(s, str(e)))
```
