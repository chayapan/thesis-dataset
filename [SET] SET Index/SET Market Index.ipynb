```python

from openpyxl import load_workbook
```

```python
fname = "SET 20200409.xlsx"
wb = load_workbook(filename = fname)
wb
```

```python
wb.worksheets
```

```python
worksheet = wb.worksheets[0]
worksheet.max_column, worksheet.max_row
```

```python
# 1223 rows

data = {}
header = []

i = 0
for r in worksheet.iter_rows():
    i += 1
    print("Row: %s" % i)
    row = []
    if i == 1: # Row 1 is header.
        for c in r:
            if c.value:
                header.append(c.value)
    else: # These are data
        for c in r:
            if c.value:
                print(c.value)
                row.append(c.value)
        data[i] = row
```

```python
data
```

```python
import pandas as pd

df = pd.DataFrame(data, index=header)
df = df.T
df
```

```python
df['Date'] = df['Date (GMT)']
df = df.drop(columns=['Date (GMT)'])
df.set_index('Date')
```

```python
df.to_csv("market-index.csv")
```
