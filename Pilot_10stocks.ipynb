```python
%matplotlib inline 
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

open_prices = s2.df[['Open']]
open_prices.sort_index(inplace=True)
open_prices
```

```python
%matplotlib inline
from matplotlib.pylab import plt

rcParams['figure.figsize'] = 15, 6
price1, = plt.plot(open_prices)

window = 100

# SMA 100
sma1 = open_prices.rolling(window).mean() # SMA 100 window=100
sma, = plt.plot(sma1, color='red')

# EMA 100
ema1 = pd.Series.ewm(open_prices, span=window).mean()  # EMA window=100
ema, = plt.plot(ema1, color='orange')

plt.legend([price1, sma, ema], ['Data', 'SMA', 'EMA'])
plt.title(s2)
```

```python
import numpy as np

def calc_returns(df):
    df['pct_change'] = df.Open.pct_change()
    df['log_ret'] = np.log(df.Open) - np.log(df.Open.shift(1))
    df['daily_return'] = (df.Open - df.Open.shift(1)) / df.Open.shift(1)
    df.dropna()
    return df


price1 = s1.df[['Open']]
price2 = s2.df[['Open']]
price3 = s3.df[['Open']]
price5 = s5.df[['Open']]
price6 = s6.df[['Open']]
price1, price2
```

```python
price1, = plt.plot(price1, color='blue')
price2, = plt.plot(price2, color='red')
price3, = plt.plot(price3)
price5, = plt.plot(price5)
price6, = plt.plot(price6)
plt.legend([price1, price2, price3, price5, price6], [s1, s2, s3, s5, s6])
```
